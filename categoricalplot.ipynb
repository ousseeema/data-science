```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
sns.barplot(data= tips , x= "smoker", y ="tip", estimator= "sum",hue="sex")
#default estimator is mean u can change it to either sum max min
```

```python
sns.boxplot(data = tips , x ="day", y = "total_bill", hue ="sex" )
```

```python
sns.countplot(x= tips["sex"], hue =tips["smoker"])
```

```python
sns.violinplot(data = tips , x ="day", y = "total_bill", hue ="sex", split=True)
```

```python
sns.stripplot(data = tips , x ="day", y = "total_bill", hue ="sex", jitter = True, dodge=True)
```

```python
sns.swarmplot(data = tips , x ="day", y = "total_bill", hue ="sex",)
```

```python
sns.violinplot(data = tips , x ="day", y = "total_bill")
sns.swarmplot(data = tips , x ="day", y = "total_bill",)
```

```python
sns.catplot(data = tips , x ="day", y = "total_bill", hue ="sex", )
```
